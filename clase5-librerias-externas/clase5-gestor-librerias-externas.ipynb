#**‚≠ê Parte 1. Gestor de Paquetes**

¬øQu√© es un gestor de paquetes en Python?

Un gestor de paquetes en Python es una herramienta que facilita la instalaci√≥n, actualizaci√≥n y eliminaci√≥n de librer√≠as y paquetes de Python. Los paquetes son colecciones de m√≥dulos de Python que se pueden usar para a√±adir funcionalidades espec√≠ficas a tus programas, sin necesidad de escribir todo el c√≥digo desde cero.

¬øPor qu√© es importante usar un gestor de paquetes?

**Facilita el acceso a bibliotecas externas:** Puedes utilizar c√≥digo desarrollado por otros para a√±adir nuevas funcionalidades a tus proyectos.

**Automatiza el proceso de instalaci√≥n:** Con un solo comando, puedes instalar todas las dependencias necesarias para tu proyecto.

**Mantiene las versiones organizadas:** Puedes gestionar diferentes versiones de paquetes para asegurarte de que tu proyecto funcione correctamente con las versiones correctas.

**Simplifica la colaboraci√≥n:** Al usar un gestor de paquetes, puedes compartir tus proyectos con otros de manera m√°s eficiente, ya que pueden replicar el entorno exacto que necesitas.

El gestor de paquetes m√°s com√∫n: `pip`

El gestor de paquetes m√°s utilizado en Python es pip. Es una herramienta que viene instalada por defecto con Python a partir de la versi√≥n 3.4. Pip permite instalar, actualizar y desinstalar paquetes de Python de manera sencilla.

Comandos b√°sicos de `pip`

- Instalar un paquete

Para instalar un paquete, se usa el comando pip install seguido del nombre del paquete. Por ejemplo, para instalar la librer√≠a pandas, que se usa para hacer peticiones HTTP, se escribir√≠a: `pip install pandas
`

- Actualizar un paquete

Para actualizar un paquete a la √∫ltima versi√≥n disponible, se usa el comando pip install --upgrade seguido del nombre del paquete. Por ejemplo:
`pip install --upgrade pandas`

Desinstalar un paquete

Para desinstalar un paquete, se usa el comando pip uninstall seguido del nombre del paquete. Por ejemplo: `pip uninstall pandas`

Listar los paquetes instalados

Para ver una lista de todos los paquetes instalados en tu entorno de Python, se usa el comando pip list: `pip list`

4.1 Grafico de Barras

Un gr√°fico de barras es una representaci√≥n gr√°fica de datos categ√≥ricos en la que las categor√≠as est√°n representadas por barras rectangulares de longitudes proporcionales a los valores que representan. Este tipo de gr√°fico es √∫til para comparar cantidades entre diferentes categor√≠as o grupos.

Caracter√≠sticas

- **Barras Rectangulares**: Las barras pueden ser verticales u horizontales.
- **Barras Verticales**: Las categor√≠as est√°n en el eje x y las frecuencias o valores est√°n en el eje y.
- **Barras Horizontales**: Las categor√≠as est√°n en el eje y y las frecuencias o valores est√°n en el eje x.
- **Altura/Longitud Proporcional**: La altura (en un gr√°fico vertical) o la longitud (en un gr√°fico horizontal) de cada barra es proporcional al valor que representa.

Categor√≠as: Representa diferentes categor√≠as o grupos que se est√°n comparando.

Etiquetas: Generalmente, cada barra tiene una etiqueta que indica la categor√≠a que representa.

4.3 Diagrama Caja Bigote (Boxplot)

Es una representaci√≥n gr√°fica que muestra la distribuci√≥n de un conjunto de `datos num√©ricos` basado en un resumen de cinco n√∫meros:

- M√≠nimo
- Q1 (1ercuartil)
- Q2 (Mediana)
- Q3 (3er cuartil)
- M√°ximo.

Este tipo de gr√°fico es √∫til para identificar la **dispersi√≥n y la asimetr√≠a** en los datos, as√≠ como para detectar valores at√≠picos (**outliers**).

Principales Caracter√≠sticas

- **Extremos de la Caja**: Los extremos de la caja representan el primer cuartil (Q1) y el tercer cuartil (Q3).
L√≠nea en la Caja: La l√≠nea dentro de la caja indica la mediana (Q2) de los datos.
- **Rango Intercuart√≠lico (IQR)**: La longitud de la caja es igual al IQR, que es la diferencia entre el tercer cuartil y el primer cuartil (IQR = Q3 - Q1). El IQR mide la dispersi√≥n de la mitad central de los datos.
Bigotes (Whiskers):

Los bigotes se extienden desde los extremos de la caja hasta los valores m√≠nimos y m√°ximos dentro de 1.5 veces el rango intercuart√≠lico (1.5 * IQR) desde los cuartiles.
Los puntos m√°s all√° de los bigotes se consideran valores at√≠picos (outliers).
Valores At√≠picos (Outliers):

Los valores que se encuentran fuera de 1.5 veces el IQR desde los cuartiles se marcan como puntos individuales y se consideran outliers (esta definici√≥n proviene de la "Prueba de Tukey").
Estos puntos proporcionan informaci√≥n sobre datos inusuales o extremos en el conjunto de datos.

```python
df['Price'].plot(kind='box', title = 'Boxplot de Precios')
```

```python
df['Price'].describe()
```
