# **Diccionarios y Tuplas. Conversiones de Tipos**

**1) Diccionarios**

Son una estructura (objeto) que almacenan una palabra clave `key` y el valor `value` correspondiente.

- Las claves son √∫nicas y pueden ser `string/int/float/tuplas` (entre otros)
- Son estructuras vers√°tiles que admiten distintos tipos de datos.
- Indexaci√≥n por rango num√©rico (se puede acceder a sus `elementos`)
- Son Mutables

1.1) Diccionario vac√≠o. Diccionario con claves `string/int`

```python
# Vac√≠o
mi_dict = {}
print(mi_dict)
```

```python
mi_dict = {
    'nombre': 'Guido',          # Clave 1 : Valor 1
    'apellido': 'Van Rossum',   # Clave 2 : Valor 2
    'edad': 50,                 # Clave 3 : Valor 3
    'programador': True         # Clave 4 : Valor 4
}
print(mi_dict)
```

```python
mi_dict = {
    1: 'Python',          # Clave 1 : Valor 1
    2: 'SQL',             # Clave 2 : Valor 2
    3: 'Java'             # Clave 3 : Valor 3
}
print(mi_dict)
```

**2.2) M√©todos para Diccionarios**

2.2.1) Obtener un elemento. (Notaci√≥n `[]` & m√©todo `get()`)

```python
# Notaci√≥n de []
dict = {
    'nombre': 'Guido',          # Clave 1 : Valor 1
    'apellido': 'Van Rossum',   # Clave 2 : Valor 2
    'edad': 50,                 # Clave 3 : Valor 3
    'programador': True         # Clave 4 : Valor 4
}
nombre = dict['nombre']
print(f'El valor de la clave "nombre" de mi diccionario es: {nombre}')
```

Nota: Es necesario conocer el nombre de la `clave`. Si busc√°ramos `dict['altura']` dar√≠a error.

```python
# M√©todo `get()`
print(dict.get('apellido'))
```

Nota: Si no encuentra el valor, la funci√≥n devuelve `False` (es posible modificar ese `valor por defecto`)

```python
print(dict.get('altura', 'Clave no encontrada'))
```

2.2.2) A√±adir o modificar un elemento: Notaci√≥n `[]`

```python
dict = {'nombre' : 'Guido',
        'altura': 1.75
        }
# Si existe, lo reemplaza
dict['altura'] = 1.80
print(dict)
```

```python
# Si no existe, la agrega
dict['peso'] = '70kg'
print(dict)
```

2.2.3) Pertenencia de una clave: `in` & `not in`

```python
dict = {'nombre' : 'Guido',
        'altura': 1.75
        }
print('nombre' in dict)
print('nombre' not in dict)
```

2.2.4) Obtener todas las claves/valores/pares
- M√©todo`keys()` para claves
- M√©todo `values()` para valores
- M√©todo `items()` para pares

```python
dict = {'nombre' : 'Guido',
        'altura': 1.75
        }
# Todas las claves: keys()
print(dict.keys())
# Todas los valores: values()
print(dict.values())
# Todos los items
print(dict.items())
```

2.2.5) Longitud de un diccionario: `len()`

```python
dict = {'nombre' : 'Guido',
        'altura': 1.75
        }
print(f'La longitud de mi diccionario es: {len(dict)}')
```

2.2.6) Combinar diccionarios: `update()`

```python
dict1 = {
    'nombre': 'Guido',
    'apellido': 'Van Rossum',
    'edad': 50,
    'programador': True
}
dict2 = {'nombre' : 'Guido Von Rossum',
        'altura': 1.75
        }
dict1.update(dict2)
print(dict1)
```

Nota: Si la clave se pisa, utiliza el del 2do diccionario (ver clave `nombre`)

2.2.7) Borrar elementos: `del()`

```python
# Borrar 1 elemento
dict = {
    'nombre': 'Guido',
    'apellido': 'Van Rossum',
    'edad': 50,
    'programador': True
}
del(dict['edad'])
print(dict)
```

```python
# Borrar TODOS los elementos: clear()
print(dict.clear())
```

2.2.8) Extraer un elemento: `pop()`

```python
dict = {
    'nombre': 'Guido',
    'apellido': 'Van Rossum',
    'edad': 50,
    'programador': True
}
dict.pop('edad')
print(dict)
```

2.2.9) Copiar un diccionario: `copy()`

```python
dict1 = {
    'nombre': 'Guido',
    'apellido': 'Van Rossum',
    'edad': 50,
    'programador': True
}
dict2 = dict1.copy()
print(dict1)
print(dict2)
```

2.1.10) Iterar `clave` y `valor` en un bucle for: `enumerate()`

```python
dict = {
    'nombre': 'Guido',
    'apellido': 'Van Rossum',
    'edad': 50,
    'programador': True
}
for posicion, clave in enumerate(dict):
    print(posicion, clave)
    valor = dict[clave]
    print(valor)
```

**2) Tuplas**

Una tupla est√° formada por un n√∫mero de valores separados por comas.

- Vers√°tiles: Admiten distintos tipos de datos
- Admiten duplicados
- INmutables (A diferencia de las `listas` y `diccionarios`)

2.1) Definir una tupla

```python
# Definici√≥n m√°s usada
tupla1 = (1, 2, 3)
tupla2 = (1, 1, 'Guido', False, 3.14)
tupla3 = (1, 'Python', [1, 3, 5], 3.14, {'Nombre': 'Guido'})
print(tupla1)
print(tupla2)
print(tupla3)
```

```python
# Definici√≥n sin par√©ntesis
tupla = 'Guido', 'Van', 'Rossum'
print(type(tupla))
print(tupla)
```

2.2) M√©todos para `tuplas`: Podemos usar los mismos que para listas salvo aquellos que modifiquen el objeto
recordemos que las tuplas son `INmutables`

Es decir, no se puede utilizar:
- `reverse()`
- `append()`
- `extend()`
- `clear()`
- `sort()`

2.2.1) Obtener un elemento: Notaci√≥n de `[]` & `index()`

```python
tupla = (1, 3.14, 'Guido', True)
# Posici√≥n
print(f'Mi primer elemento de la tupla es: {tupla[0]}')
# Indexaci√≥n
print(f'El √≠ndice de mi elemento 1 es {tupla.index(1)}')
```

2.2.2) Contar apariciones de un elemento: `count(elemento)`

```python
tupla = (1, 2, 3, 3, 4, 5, 2, 2, 4, 2, 5, 2)
print(tupla.count(2))
```

# Funciones üßÆ

Estructura que nos permite agrupar c√≥digo para no repetir y poder reutilizarlo.
Est√° b√°sicamente compuesta por:
- `nombres` de la funci√≥n
- `parametros` a recibir (Luego el valor real pasado a la funci√≥n se denomina `argumento`
- `return`: Valor de retorno

Para llamar a la funci√≥n se utiliza el nombre seguido de par√©ntesis (agreg√°ndole los par√°metros en caso de que reciba):

`funci√≥n(argumento1, argumento2)`

Notas:
- Prestar atenci√≥n a la `indentaci√≥n` para reflejar el c√≥digo que est√° "dentro" de nuestra funci√≥n.
- Es importante documentar la acci√≥n realizada por la funci√≥n al definirla.


**1) Definicion de funci√≥n**

- Funci√≥n que no realiza acci√≥n
- Funci√≥n que recibe un par√°metro
- Funci√≥n que recibe un par√°metro y devuelve un resultado

1.1) Funcion que no realiza acci√≥n: `pass`

```python
# Defino funci√≥n
def funcion1():
    """Funci√≥n que no realiza ninguna acci√≥n
    """
    pass

# Llamo a mi funci√≥n
funcion1()
```

Nota: ¬øY para qu√© es √∫til si no hace nada? Muchas veces cuando programamos, vamos armando la estructura de nuestra aplicaci√≥n y definiendo las funciones y acciones que deber√°n realizar pero sin codearla en ese momento, con lo cual terminamos de escribirlas en un paso posterior.

1.2) Funci√≥n que recibe un par√°metro

```python
# Defino funci√≥n
def funcion2(nombre):
    """Funci√≥n que recibe como par√°metro un nombre e imprime en pantalla un mensaje
    """
    print(f'Hola {nombre}!')

# Llamo a mi funci√≥n
funcion2('Luciano')
```

1.3) Funci√≥n que recibe 2 par√°metros y retorna un valor.

```python
# Defino funci√≥n
def funcion3(nombre, edad):
    """ Funci√≥n que recibe parametros de nombre, edad y retorna un mensaje con
    la categor√≠a a la que corresponde la persona
    """
    if edad < 8:
      categoria = 'inferiores'
      mensaje = f'{nombre} est√°s en la categor√≠a: {categoria}'
    elif 8 < edad < 12:
      categoria = 'intermedia'
      mensaje = f'{nombre} est√°s en la categor√≠a: {categoria}'
    else:
      categoria = 'superior'
      mensaje = f'{nombre} est√°s en la categor√≠a: {categoria}'

    return mensaje

# Llamo a la funci√≥n y almaceno el resultado del 'return' en una variable `mensaje`
mensaje = funcion3('Luciano', 32)
# Imprimo mi variable `mensaje` que captur√≥ el `return` de mi funcion3
print(mensaje)
```

1.3) Par√°metros por defecto

En caso de que no se proporcione valor al `argumento` en la llamada a la `funci√≥n`, el parametro correspondiente puede asignarse con un valor predefinido `por defecto` al definila.

```python
def caracteristicas_persona(nombre, apellido, comida_favorita = 'Pizza'):
    """Funci√≥n que recibe de parametros nombre, apellido y comida_favorita e
    imprime un mensaje en pantalla
    """
    print(f'Hola {nombre} {apellido}, tu comida favorita es: {comida_favorita}')

# LLamo a mi funci√≥n sin pasar el par√°metro ``comida_favorita`
caracteristicas_persona('Luciano', 'Delgado')
# LLamo a mi funci√≥n `pasando el par√°metro ``comida_favorita` (pisa par√°metro por defecto)
caracteristicas_persona('Luciano', 'Delgado', 'Tarta de acelga')
```

```python
def saludo(
    nombre:str      # Ac√° sugiero que el argumento a recibir deber√≠a ser `string`
    ):
  """ Funcion que recibe el parametro nombre (en el cual se sugiere que sea string)
  """
  print(f'Hola {nombre}')

# LLamo a mi funci√≥n
saludo('Luciano')
```

**2) Argumentos de una funci√≥n**

2.1) Argumentos `posicionales`

Se copian en sus correspondientes par√°metros **en orden** (durante el llamado a la funci√≥n, no impacta en su definici√≥n)

Nota: Se necesita recordar el orden de los par√°metros.

```python
def armar_pc(proveedor, nucleos, frecuencia):
    """Funcion que recibe como parametros proveedor, cantidad de nubleos y frecuencia
    y retorna un diccionario con las caracter√≠sticas de mi PC.
    """
    dict = {
        "proveedor": proveedor,
        "nucleos": nucleos,
        "frecuencia": frecuencia,
        }

    return dict
```

```python
# LLamado a la funci√≥n pasando argumentos posicionales
resultado = armar_pc(
    'Intel',
    4,
    '80MHz'
    )
print(f'Resultado de llamada a mi funci√≥n "armar_pc" pasando parametros posicionales: \n{resultado}')
```

2.2) Argumentos `nominales`

Se asignan **por nombre** a cada par√°metro (no impactan en la definici√≥n de la funci√≥n)

Nota: No se necesita recordar el posicionamiento al llamar a la funci√≥n.

```python
resultado = armar_pc(
    proveedor='Intel',
    nucleos=4,
    frecuencia='80MHz')

print(f'Resultado de llamada a mi funci√≥n "armar_pc" pasando parametros nominales es:\n {resultado}')
```

2.3) Argumentos `mixtos`
 Los posicionales deben ir antes que los nominales, sino Python no tendr√≠a forma de discernir a qu√© par√°metro corresponde cada argumento.

```python
# Reutilizando la funci√≥n armar_pc definida previamente
armar_pc(
    'Intel',
    nucleos=4,
    frecuencia='80MHz'
    )
```

El siguiente c√≥digo dar√≠a `error` ya que estoy pasando como argumento `posicional` posterior a uno `ordinal`.

**3) Packing & Unpacking**

Son formas de capturar el resultado del `return` (Valor de retorno)de una funci√≥n.

3.1) Packing

Ocurre cuando una `funci√≥n` devuelve m√∫ltiples valores y los capturamos todos juntos en una **√∫nica** `variable`.

3.1.2) Ejemplo 1: retornando valores de mismo tipo

```python
def obtener_puntos():
    """Funci√≥n que devuelve dos valores enteros fijos
    """
    return 10, 20  # Retorna dos valores: 10 y 20

# Llamado con packing:
resultado1 = obtener_puntos()  # Capturamos los valores retornados en la variable 'puntos'
print(resultado1)
```

3.1.2) Ejemplo 2: retornando valores de distinto tipo

```python
def obtener_condicion(condicion_binaria: str):
    """ Funcion que recibe un par√°metro sugerido del tipo 'string'
    y devuelve
    """
    if condicion_binaria == 'Yes':
      return True, 1
    elif condicion_binaria == 'No':
      return False, 0
    else:
      return 'no condition'

# Llamado con packing:
resultado2 = obtener_condicion('Yes')
print(resultado2)
```

Nota: Para el ejemplo retornamos 2 valores pero no hay l√≠mites, podr√≠an ser 3,4,5... etc

¬øC√≥mo accedo a cada valor luego del `packing`?
Utilizamos referencia por √≠ndice en nuestra variable empaquetada.

```python
primer_valor_retornado = resultado2[0]
segundo_valor_retornado = resultado2[1]
print(f'Mi primer valor retornado fue: {primer_valor_retornado}')
print(f'Mi segundo valor retornado fue: {segundo_valor_retornado}')
```

3.2) Unpacking

Ocurre cuando una `funci√≥n` devuelve m√∫ltiples valores y los capturamos todos juntos en una **varias** `variables`.

3.2.1) Ejemplo 1: Devolviendo `strings`

Definir una funci√≥n `listar_articulos_a` que reciba como parametros una lista de articulos.
- Aclarar que se espera un objeto de tipo `lista`.
- La funci√≥n debe contener un bucle `for` que recorra cada art√≠culo y lo almacene en una lista en caso de comenzar con la letra a, independientemente de si el dato viene en may√∫scula o min√∫scula.

Nota: La `lista` que se usar√° para almacenar valores debe definirte previo a la funci√≥n.

```python
# Definimos una lista donde almacenar nuestras cotizaciones
lista = []
def listar_articulos_a(lista_articulos: list):
    """Funcion que recibe como parametro el nombre de un articulo (string) y lo
    anexa a una lista si cumple con la condici√≥n
    """
    for articulo in lista_articulos:
      primer_letra = articulo.lower().startswith("a")
      if primer_letra == True:
        lista.append(articulo)

    return lista

# Llamo a mi funci√≥n
lista = listar_articulos_a(['Adaptador','Celular', 'Armario', 'Heladera','Auto',' Anana'])
print(lista)
```

- `('HTML', 'Javascript', 'CSS', 'React', 'Django','Flask')`: Stack orientado a FrontEnd developer
- `('PowerBI', 'Tableau', 'LookerStudio', 'Looker',)`: Stack orientado a Analytics

```python
mi_dict_1 = {'nombre': 'Luciano', 'apellido': 'Delgado', 'edad': 32, 'stack': 'Python'}
mi_dict_2 = {'nombre': 'Franco', 'apellido': 'Di Leo', 'edad': 30, 'stack': 'Javascript'}
mi_dict_3 = {'nombre': 'Cecilia', 'apellido': 'Enrich', 'edad': 34, 'stack': 'PowerBI'}

def f1(diccionario):
  """Funcion que recibe como parametro un diccionario y devuelve el nombre y el stack al que
  est√° orientado la persona
  """
  for clave, valor in diccionario.items():
      # print(valor)
      if clave == 'nombre':
        nombre = valor
      if clave == 'stack' and valor in ('Python', 'SQL', 'MySQL', 'BigQuery', 'Spark' ):
          return nombre,'Stack orientado a Data Scientist'
      elif clave == 'stack' and valor in ('HTML', 'Javascript', 'CSS', 'React', 'Django','Flask'):
          return nombre, 'Stack orientado a FrontEnd developer'
      elif clave == 'stack' and valor in ('PowerBI', 'Tableau', 'LookerStudio', 'Looker',):
          return nombre,'Stack orientado a Analytics'
      else:
        continue

resultado1 = f1(mi_dict_1)
resultado2 = f1(mi_dict_2)
resultado3 = f1(mi_dict_3)
print(resultado1)
print(resultado2)
print(resultado3)
```

ü´¥üèª Parte 2 - Ejercicio de tarea 2

Definir una funci√≥n que reciba 2 par√°metros `int` y realice la multiplicaci√≥n de ellos.

Luego se deben clafificar como `positivo/negativo/nulo` y retornar el resultado de la multiplicacion junto con la clasificaci√≥n.

Capturar el `return` de la funci√≥n a trav√©s de la forma `unpacking`

```python
def f2(a:int, b:int):
    """Funcion que recibe como parametro 2 valores enteros y devueve la multiplicacion
    junto con la clasificacion
    """
    mult = a*b
    if mult >0:
      return mult, 'Positivo'
    elif mult <0:
      return mult, 'Negativo'
    elif mult ==0:
      return mult, 'Nulo'

# Llamo a mi funci√≥n con unpacking
resultado, clasificacion = f2(4,0)
print(resultado, clasificacion)
resultado, clasificacion = f2(5,3)
print(resultado, clasificacion)
resultado, clasificacion = f2(-2,3)
print(resultado, clasificacion)
```
