**1.2) Otros condicionales: `match/case`**

Este algoritmo nos sirve cuando los condicionales a evaluar no son demasiado complejos. Otorga una sintaxis m√°s legible y sencilla.

```python
x = 10
match x:
    case 1:
        print('x es 1')
    case 5:
        print('x es 5')
    case _:
        print('x no esta dentro de mis valores esperados <unknown>')
```

Vamos con un ejemplo m√°s complejo.

Nota 1: El caracter `|` es la representaci√≥n del condicional `or`

Nota 2: El caracter `_` es una convenci√≥n en python para representar "cualquier otro caso".

```python
edad = int(input('Ingrese su edad: '))
match edad:
        case 0 | 1 | 2:
            print("Beb√©")
        case 3 | 4 | 5:
            print("Ni√±o peque√±o")
        case 6 | 7 | 8 | 9 | 10:
            print("Ni√±o")
        case 11 | 12 | 13 | 14 | 15:
            print("Adolescente")
        case 16 | 17 | 18 | 19 | 20:
            print("Joven adulto")
        case _:
            print("Adulto")
```

Escribe un programa en Python que solicite al usuario que ingrese un n√∫mero.
- Luego, el programa debe determinar si el n√∫mero ingresado es par o impar, y realizar una transformaci√≥n sobre el n√∫mero seg√∫n el caso.
  - Si el n√∫mero es par, dividirlo por 2.
  - Si el n√∫mero es impar, elevarlo al cuadrado.
Finalmente, el programa debe imprimir el resultado transformado en pantalla.

```python
# Pedir al usuario que ingrese un n√∫mero
numero = int(input("Por favor, ingresa un n√∫mero: "))

# Verificar si el n√∫mero ingresado es par o impar. Recordemos que % se usa para obtener el 'resto' de una divisi√≥n
if numero % 2 == 0:
    # Si el n√∫mero es par, dividirlo por 2 utiliando '//' para divisi√≥n entera
    resultado_transformado = numero // 2
else:
    # Si el n√∫mero es impar, elevarlo al cuadrado
    resultado_transformado = numero ** 2

print(f"El n√∫mero transformado es {resultado_transformado}.")
```

3.1.3) Uso del `continue` para condiciones anidadas

Sin `continue`

```python
for i in range(5):
    if i % 2 == 0:
        print(f"{i} es par")
    print("Esta l√≠nea se ejecuta siempre")  # Sin el continue: La ejecuci√≥n llega hasta esta l√≠nea.
```

Con `continue`

```python
for i in range(5):
    if i % 2 == 0:
        print(f"{i} es par")
        continue
    print("Esta l√≠nea se ejecuta solo si i no es par") # Con el continue: La ejecuci√≥n NO llega hasta esta l√≠nea (contin√∫a la siguiente vuelta).
```
